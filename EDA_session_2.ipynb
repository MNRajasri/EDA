### create dataframe using dict

```python
name=["avinash","aakash","aadhya"]
age=[25,30,35]
pd.DataFrame(zip(name,age),columns=["name","age"])
```

- automatically keys become columns

```python
dict1={"names":["avinash","aakash","aadhya"],"ages":[25,30,35]}
pd.DataFrame(dict1)
```

```python
dict2={"names":["avinash"],"ages":[25],"city":["hyd"]}
df=pd.DataFrame(dict2)
df
```

```python
dict2={"names":"avinash","ages":25,"city":"hyd"}
pd.DataFrame(dict2)
```

- If using all scalar values, you must pass an index

```python
dict2={"names":"avinash","ages":25,"city":"hyd"}
df=pd.DataFrame(dict2,index=[1,2,3])
df
```

### insert a col at particular loc

```python
name=["avinash","aakash","aadhya"]
age=[25,30,35]
df=pd.DataFrame(zip(name,age),columns=["name","age"])
df
```

```python
df.insert(1,"id",[100,200,300])
df
```

### adding a new row

```python
dict2={"names":"avinash","ages":25,"city":"hyd"}
df=pd.DataFrame(dict2,index=[1,2,3])
df
```

```python
df.iloc[len(df)+1]={"names":"avinash","ages":25,"city":"hyd"}
df
```

```python
df.loc[len(df)+1]={"names":["avinash"],"ages":[25],"city":["hyd"]}
df
```

```python
df.loc[len(df)]={"names":"avinash","ages":25,"city":"hyd"}
df
```
