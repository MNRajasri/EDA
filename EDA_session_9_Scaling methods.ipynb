### normalization

```python
visa_df=pd.read_csv(r"C:\Users\Rajasri\Documents\DATA SCIENCE Naresh IT\Data files\Visadataset - Visadataset.csv")
visa_df.head(2)
```

```python
val=visa_df["prevailing_wage"]
minn=min(visa_df["prevailing_wage"])
maxx=max(visa_df["prevailing_wage"])
Nr=(val-minn)
Dr=(maxx-minn)
minmax=Nr/Dr
visa_df["prevailing_wage_mm"]=minmax
visa_df
```

- minmaxscalar

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
visa_df["prevailing_wage_mmp"]=mm.fit_transform(visa_df[["prevailing_wage"]])
visa_df
```
