```python
dict3=dict(visa_df.dtypes)

categorical_data=[]
numerical_data=[]
for key in dict3:
    if dict3[key] == "object":
        categorical_data.append(key)
    else:
        numerical_data.append(key)

print("Categorical data:",categorical_data)
print("Numerical data:",numerical_data)
```

```python
dict3=dict(visa_df.dtypes)

categorical_data=[key for key in dict3 if dict3[key] == "object"]
numerical_data=[]
for key in dict3:
    if dict3[key] == "object":
        categorical_data.append(key)
    else:
        numerical_data.append(key)

print("Categorical data:",categorical_data)
print("Numerical data:",numerical_data)
```
